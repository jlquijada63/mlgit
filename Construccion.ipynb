# Construccion del DataFrame

Los DataFrames de pandas pueden construirse de dos formas:
1. **Constructor**: `pd.DataFrame()`, que recibe como parametro una secuencia y otros parametros adicionales como el nombre de las columnas, la columna que se utiliza de indice etc.
2. **Importacion de archivos**: Esta es la forma mas frecuente en la practica real. Pandas tiene funciones para importar diferentes tipos de archivos. Los mas frecuentes son archivos del tipo .csv

## Creaci√≥n con el constructor

```python
import pandas as pd
import numpy as np
```

```python
cls = 'manzanas naranjas peras melocotones'
df = pd.DataFrame(np.reshape(np.arange(80),(20,4)),columns=cls.split())
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
# eliminamos la columna Name que no nos sirve para el analisis
df = df.drop('Name',axis=1)
```

```python
df.head()

```

```python
# Cambiamos los nombres de las columnas por su traduccion al castellano
lista = 'sobrevive clase sexo edad familiares hijos tarifa'
df.columns = lista.split()
```

```python
df.head()
```

```python
df.tail()
```

```python
new_index = [element for element in range(1,df.shape[0],1)]
df = df.reindex(new_index)
```

```python
df.head()
```

```python
# La variable sexo es de tipo string y hay que pasarla a numerica
```

```python
df['sexo_bin'] = df.sexo == 'female'
```

```python
df.head()
```

```python
df['sexo_b']=df['sexo_bin'].astype('int',copy=True)
```

```python
df.head()
```

```python
#eliminamos la columna sexo y sexo_bin
```

```python
df = df.drop(['sexo','sexo_bin'],axis=1)
```

```python
# renombramos la columna sexo_b
```

```python
df.rename({'sexo_b':'sexo'},axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#vemos como es la distribucion por sexos
df.groupby('sexo').mean()
```
